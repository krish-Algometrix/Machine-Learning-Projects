```python
karate_dataset = nx.read_edgelist('kac_data.txt', nodetype=str)
```

```python
nx.info(karate_dataset)
```

Visualizing the networtk of karate club members 

```python
nx.draw_networkx(karate_dataset)
```

```python
# measuring the Density of the Karate Club network which is a undirected graph
density_karate_club = nx.density(karate_dataset)
print(density_karate_club)
```

```python
# degree centrality of the network
kcentrality= nx.degree_centrality(karate_dataset)
kcentrality
```

```python
max_key = max(kcentrality, key = lambda x: kcentrality[x])
max_key
# The member 33 is the member to whom all other karate club members are connected.
```

```python
nx.clustering(karate_dataset)
# we clearly find three clusters of karate_member groups .
```

```python
# cliques 
list(nx.find_cliques(karate_dataset))
```

```python
nx.eccentricity(karate_dataset) # the distance of each node from the zero node.
```

```python
sim = nx.simrank_similarity(karate_dataset, source=None, target=None, importance_factor=0.9, max_iterations=1000, tolerance=0.0001)
```

```python
import numpy as np
sim_array = np.array([[sim[u][v] for v in karate_dataset ]for u in karate_dataset])
sim_array
```

Analysis of karate club network relationships reveals that primarily there are clear groupings among the karate club members which may be due to experience/earlier batches and also specific skills and seniority.

The clustering of the network results in clearly at least three major groups.
